```python
# import dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io

df = pd.read_csv(io.BytesIO(uploaded['mushrooms.csv']))
```

```python
# check dataframe
df.head()
```

```python
# get column types, counts and column names
df.info()
```

```python
# drop column that isn't needed for analysis
new_df = df.drop(columns = ["veil-color"], axis=1)
```

```python
# get value counts for y
new_df['class'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
new_df = pd.get_dummies(new_df)

new_df.head()
```

```python
# split our data into our features and target arrays
X = new_df.drop(columns = ['class_e', 'class_p'], axis =1)
y = new_df[['class_e']]
```

```python
# split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=23)
```

```python
# create a StandardScaler instances
scaler = StandardScaler()

# fit the StandardScaler
X_scaler = scaler.fit(X_train)

# scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=6, activation="relu", input_dim=113))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=4, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
